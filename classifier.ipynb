```python
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
import PIL
```

```python
dataset = Path('./Rice_Image_Dataset')

image_count = 0

for subdir in dataset.iterdir():
    if subdir.is_dir():
        c = sum(1 for x in subdir.glob('*') if x.is_file())
        image_count += c
        print(f'{subdir}:\t{c} images')

print(f'\n{image_count} images in total')
```

```python
arborio = list(dataset.glob('Arborio/*'))
basmati = list(dataset.glob('Basmati/*'))
ipsala = list(dataset.glob('Ipsala/*'))
jasmine = list(dataset.glob('Jasmine/*'))
karcadag = list(dataset.glob('Karacadag/*'))

plt.subplot(3, 3, 1)
plt.imshow(PIL.Image.open(str(arborio[1])))
plt.axis('off')
plt.title('Arborio')

plt.subplot(3, 3, 3)
plt.imshow(PIL.Image.open(str(basmati[1])))
plt.axis('off')
plt.title('Basmati')

plt.subplot(3, 3, 5)
plt.imshow(PIL.Image.open(str(ipsala[1])))
plt.axis('off')
plt.title('Ipsala')

plt.subplot(3, 3, 7)
plt.imshow(PIL.Image.open(str(jasmine[1])))
plt.axis('off')
plt.title('Jasmine')

plt.subplot(3, 3, 9)
plt.imshow(PIL.Image.open(str(karcadag[1])))
plt.axis('off')
plt.title('Karacadag')

plt.show()
```

```python
raw_images = []
labels = []

for path in arborio[:5] + basmati[:5] + ipsala[:5] + jasmine[:5] + karcadag[:5]:
    raw_image = PIL.Image.open(str(path)).getdata()
    label = str(path).split('/')[1]
    raw_images.append(raw_image)
    labels.append(label)

raw_images = np.array(raw_images)
labels = np.array(labels)
```
